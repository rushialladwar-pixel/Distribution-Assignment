## Q1: Probability of rolling a 3 exactly 5 times in 30 rolls

Using Binomial Distribution:
- n = 30 trials
- p = 1/6 (probability of rolling a 3)
- x = 5

Excel Formula:
`=BINOM.DIST(5,30,1/6,FALSE)`

```

import math
from math import comb

n = 30
x = 5
p = 1/6

prob = comb(n, x) * (p**x) * ((1-p)**(n-x))
prob
```

## Q2: Uniform Distribution using RAND()

Excel: `=RAND()` generates values between 0 and 1.

The histogram will be **uniform (flat)**, meaning all values have equal probability.

```

import numpy as np
import matplotlib.pyplot as plt

data = np.random.rand(100)

plt.hist(data, bins=10)
plt.title("Uniform Distribution Histogram")
plt.show()
```

## Q4: Standardization (Z-score)

Formula:
Z = (X - μ) / σ

Purpose:
- Converts data to common scale
- Mean becomes 0
- Standard deviation becomes 1
- Helps compare different datasets

## Q7: Probability of at least 8 successes in 15 trials (p=0.5)

Excel:
`=1 - BINOM.DIST(7,15,0.5,TRUE)`

```

from scipy.stats import binom

prob_at_least_8 = 1 - binom.cdf(7, 15, 0.5)
prob_at_least_8
```
